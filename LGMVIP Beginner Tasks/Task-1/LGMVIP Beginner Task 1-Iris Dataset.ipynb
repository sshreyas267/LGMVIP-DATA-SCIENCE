# Training of Data

```python
X=df[['ID','sepal length(cm)' , 'sepal width(cm)' , 'petal length(cm)','petal width(cm)']]
y=df['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state = 0)
```

# Linear Regression


```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
model= LogisticRegression(solver='lbfgs', max_iter=1000)
model.fit(X_train, y_train)
y_prod=model.predict(X_test)
print('Accuracy of Logistic regression classifier on iris dataset: {:f}' .format(model.score(X_test, y_test)))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_prod))
```

# Accuracy of Model

```python
from sklearn.metrics import accuracy_score
train_score=str(model.score(X_train, y_train)*100)
test_score=str(model.score(X_train, y_train)*100)
accu_score=str(model.score(X_train, y_train)*100)
print(f'Training Score {train_score[:6]}%\nTest Score {test_score[:6]}%\nAccuracy Score {accu_score[:6]}%')
```
